```python
import pandas as pd
import numpy as np
```

```python
df_11 = pd.read_csv('school_sum11_final.csv')
df_12 = pd.read_csv('school_sum12_final.csv')
df_13 = pd.read_csv('school_sum13_final.csv')

df_14 = pd.read_csv('school_sum14_final.csv')
df_15 = pd.read_csv('school_sum15_final.csv')
df_16 = pd.read_csv('school_sum16_final.csv')
df_17 = pd.read_csv('school_sum17_mod.csv')
df_18 = pd.read_csv('school_sum18_final.csv')
df_19 = pd.read_csv('school_sum19_final.csv')
```

```python
df_11 = df_11.rename(columns={'zip ': 'zip'})
df_12 = df_12.rename(columns={'zip ': 'zip'})
df_13 = df_13.rename(columns={'Zip': 'zip'})
df_14 = df_14.rename(columns={'Zip': 'zip'})
df_15 = df_15.rename(columns={'Zip': 'zip'})
df_16 = df_16.rename(columns={'Zip': 'zip'})
df_17 = df_17.rename(columns={'Zip': 'zip'})
df_18 = df_18.rename(columns={'Zip': 'zip'})
df_19 = df_19.rename(columns={'Zip': 'zip'})
df_14 = df_14.rename(columns={'Total': 'Year_2014'})
df_17 = df_17.rename(columns={'Total': 'Year_2017'})
```

```python
df_11 = df_11.astype('int64')
df_12 = df_12.astype('int64')
df_13 = df_13.astype('int64')
df_14 = df_14.astype('int64')
df_15 = df_15.astype('int64')
df_16 = df_16.astype('int64')
df_17 = df_17.astype('int64')
df_18 = df_18.astype('int64')
df_19 = df_19.astype('int64')
```

```python
merged_df = df_11.merge(df_12, on='zip') \
    .merge(df_13, on='zip') \
    .merge(df_14, on='zip') \
    .merge(df_15, on='zip') \
    .merge(df_16, on='zip') \
    .merge(df_17, on='zip') \
    .merge(df_18, on='zip') \
    .merge(df_19, on='zip')
```

```python
merged_df.to_csv('US_IL_Education.csv', index=False)
```
